```python
import numpy as np
```

```python
# create an array composed of 1000 values
points = np.arange(-5,5,0.01)
print('points: \n',points)
points_10 = np.arange(-10,10,1)
print('points_10: \n',points_10)
```

```python
# meshgrid es una funcion que crea 2 matrices a partir de un vector
# comun y repiten sus valores para que den todos los posibles pares.
xs, ys = np.meshgrid(points,points)
print('xs: \n',xs, '\n\nys: \n',ys)

# points_10
xs10, ys10 = np.meshgrid(points_10,points_10)
print('xs10: \n',xs10, '\n\nys10: \n',ys10)
```

```python
# z seria la distancia entre el punto x y el punto x del par
z = np.sqrt(xs10**2 + ys10**2) # solo con el vector points_10
print(z.round(1))
```

```python
import matplotlib.pyplot as plt
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```
