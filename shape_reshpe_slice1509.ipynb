```python
arr2 = np.arange(1,364+1)
arr2_shape = np.shape(arr2)
print("arr2_shape: ", arr2_shape)
arr2
```

```python
arr2_reshaped_2d = np.reshape(arr2,(4,91))
arr2_reshaped_2d
```

```python
# indexing
print(arr2_reshaped_2d[3,17])
```

```python
# slicing

sliced_array = arr2_reshaped_2d[2:]
sliced_array
```

```python
# extract a 1d array from the 2d array
sliced_array2 = sliced_array[1]
sliced_array2
```

```python
# aggregation functions

# np.sum()
sum_all_elements = np.sum(sliced_array2)
print("The sum of all elements of sliced_array2 is: ",sum_all_elements)

#np.mean()
mean = np.mean(sliced_array2)
print("The mean of all elements of sliced_array2 is: ",mean)

#median()
median = np.median(sliced_array2)
print("The median of all elements of sliced_array2 is: ",median)

# np.std()
std = np.std(sliced_array2)
print("The standard deviation of all elements of sliced_array2 is: ",std)
```
