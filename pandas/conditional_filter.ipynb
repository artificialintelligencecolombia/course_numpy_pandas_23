```python
import pandas as pd
import numpy as np
```

```python
# Setting a random seed for reproducibility
np.random.seed(0)

# Creating a random DataFrame with 13 rows and 10 columns
# Let's assume this data represents a survey of a sports club
df = pd.DataFrame(np.random.randint(1, 100, size=(13, 10)), 
                  columns=['Age', 'Height_cm', 'Weight_kg', 'RunningTime_min', 'SwimmingTime_min', 
                           'CyclingTime_min', 'GymHours_week', 'YogaHours_week', 'BasketballHours_week', 'SoccerHours_week'])

df.head()  # Display the first few rows of the DataFrame

```

```python
# conditional filtering

age_more_than_50 = df[df['Age'] > 50] # df['Age'] < 50: prints a boolean indicator, then the df[..]
# prints the True values of the conditional

age_different_than_50 = df[~(df['Age']== 82)] # ~ is symbol for not

print(age_more_than_50)
print('\n\n',age_different_than_50)
```
